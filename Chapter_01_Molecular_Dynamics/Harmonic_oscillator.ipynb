```python
import numpy as np 
from matplotlib import pyplot as plt

N = 2000
dt = 0.1
k = 0.1

x = np.zeros(N)
V = np.ones(N)

for t in range(N-1):
    x[t+1] = x[t] + V[t]*dt                 # Harmonic oscillator position equation, Euler
    V[t+1] = V[t] - k*x[t]*dt               # Harmonic oscillator velocity equation, Euler

plt.figure(figsize=(15,5))
plt.plot(x)


x  = np.zeros(N)
xh = np.zeros(N)
V  = np.ones(N)

for t in range(N-1):
    xh[t] = x[t] + 0.5*V[t]*dt      # Harm. osc. position eq., intermediate step, leapfrog
    V[t+1] = V[t] - k*xh[t]*dt      # Harm. osc. velocity eq., leapfrog
    x[t+1] = xh[t] + 0.5*V[t+1]*dt  # Harm. osc. position eq., final step, leapfrog

plt.plot(x)
plt.legend(['Euler','Leap-frog'])
plt.ylabel('$x$')
plt.xlabel('$t$')
plt.title('Harmonic oscillator')
plt.rcParams.update({'font.size': 18})
```
