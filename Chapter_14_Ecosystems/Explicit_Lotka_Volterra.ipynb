```python
# Explicit Lotka-Volterra Model
import numpy as np 
from matplotlib import pyplot as plt

N = 400             # Simulation length

x0 = 1
y0 = 1
alpha = 0.67
beta = 1.33
gamma = 1
delta = 1
dt = 0.05

x = np.zeros(N)
y = np.zeros(N)
x[0] = x0
y[0] = y0

for t in range(N-1):
    x[t+1] = x[t] + (alpha*x[t]-beta*x[t]*y[t])*dt
    y[t+1] = y[t] + (delta*x[t]*y[t]-gamma*y[t])*dt

fig = plt.figure(figsize=(18,5))
gs = fig.add_gridspec(3,12)

plt.subplot(gs[:,0:3])
plt.plot(x)
plt.plot(y)
plt.legend(['Preys','Predators'])
plt.ylabel('$x,y$')
plt.xlabel('$t$')



plt.subplot(gs[:,4:7])
I = delta*x-gamma*np.log(x) + beta*y - alpha*np.log(y)
plt.plot(I)
plt.plot([0,N], I[0]*np.ones(2), '--')
plt.ylabel('$I(t)$')
plt.xlabel('$t$')
plt.ylim([0,3.9])

plt.subplot(gs[:,8:11])
plt.plot(x,y)
plt.plot(x[0],y[0],'s', markersize=12)
plt.plot(x[-1],y[-1],'o', markersize=12)
plt.ylabel('$y$')
plt.xlabel('$x$')

plt.rcParams.update({'font.size': 18})
plt.suptitle('Explicit Lotka-Volterra Model')
```
